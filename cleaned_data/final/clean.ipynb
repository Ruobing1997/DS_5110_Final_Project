```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("1.csv")
df2 = pd.read_csv("2.csv")
df3 = pd.read_csv("3.csv")
```

```python
df2["event_time"] = pd.to_datetime(df2["event_date"], errors="coerce")
df2["victim_profession"] = df2["occupational_category_of_person_affected"]
df2["perpetrator_type"] = df2["aggressor"]
df2["violence_type"] = df2["type_of_violence"]
df2["severity"] = df2["severity_of_assault"]
df2["emotional_impact"] = df2["emotional_and__or_psychological_impact"]
df2["assault_desc"] = df2["primary_assault_description"]
df2["response_action"] = df2["response_action_taken"]
df2["contributing_factors"] = df2["primary_contributing_factors"]
df2["department"] = df2["facility_type"] 

def clean_violence_type(v):
    if pd.isna(v):
        return np.nan
    val = v.strip().lower()
    if "physical" in val:
        return "physical"
    elif "verbal" in val:
        return "verbal"
    elif "threat" in val:
        return "threat"
    else:
        return "other"
df2["violence_type"] = df2["violence_type"].apply(clean_violence_type)

def clean_severity(s):
    if pd.isna(s):
        return np.nan
    val = s.strip().lower()
    if "mild" in val:
        return "Mild - Mild Soreness/Abrasions/Scratches/Small Bruises"
    elif "moderate" in val:
        return "Moderate - Major Soreness/Cuts/Large Bruises"
    elif "severe" in val:
        return "Severe"
    elif "none" in val or "no contact" in val:
        return "None - No Contact or Unwanted Contact w/o Injury"
    else:
        return "Unknown"
df2["severity"] = df2["severity"].apply(clean_severity)

def clean_perp_type(p):
    if pd.isna(p):
        return np.nan
    p = p.strip().lower()
    if p in ["inpatient", "outpatient", "patient", "resident (ltc)"]:
        return "patient"
    elif p in ["employee","affiliate (contracted staff, student)"]:
        return "employee"
    elif p == "visitor":
        return "visitor"
    else:
        return "other"
df2["perpetrator_type"] = df2["perpetrator_type"].apply(clean_perp_type)

df2["department"] = df2["department"].fillna("Unknown").str.strip().str.lower()

wpv_cols = [c for c in df2.columns if c.startswith("wpv_")]

print("wpv columns found in df2:", wpv_cols)

def unify_violence_type_wpv(v):
    if pd.isna(v):
        return np.nan
    val = str(v).strip().lower()
    if "physical" in val:
        return "physical"
    elif "verbal" in val:
        return "verbal"
    elif "threat" in val:
        return "threat"
    else:
        return "other"

if "wpv_event_type" in df2.columns:
    df2["violence_type"] = df2["violence_type"].fillna("")
    df2["wpv_event_type_clean"] = df2["wpv_event_type"].apply(unify_violence_type_wpv)

    def merge_violence(row):
        if row["violence_type"] in ["", "other", np.nan]:
            return row["wpv_event_type_clean"]
        else:
            return row["violence_type"]
    df2["violence_type"] = df2.apply(merge_violence, axis=1)

    df2.drop(columns=["wpv_event_type_clean"], inplace=True)

if "wpv_psych_impact" in df2.columns:
    df2["emotional_impact"] = df2["emotional_impact"].fillna("")
    df2["wpv_psych_impact"] = df2["wpv_psych_impact"].fillna("")
    def merge_emotion(row):
        if row["emotional_impact"] == "":
            return row["wpv_psych_impact"]
        else:
            return row["emotional_impact"]
    df2["emotional_impact"] = df2.apply(merge_emotion, axis=1)

if "wpv_staff_immediate_action" in df2.columns:
    df2["response_action"] = df2["response_action"].fillna("")
    df2["wpv_staff_immediate_action"] = df2["wpv_staff_immediate_action"].fillna("")
    def merge_action(row):
        if row["response_action"] == "":
            return row["wpv_staff_immediate_action"]
        else:
            # 拼接
            return row["response_action"] + "; " + row["wpv_staff_immediate_action"] if row["wpv_staff_immediate_action"] else row["response_action"]
    df2["response_action"] = df2.apply(merge_action, axis=1)
```

```python
print("=== df3 Info ===")
df3.info()
df3.head()
```

```python
df3['event_time'] = pd.to_datetime(df3['event_time'], errors='coerce')
df3['victim_profession'] = df3.apply(
    lambda x: x['victim_profession'] 
              if pd.isna(x['victim_primary_job']) 
              else f"{x['victim_profession']}, {x['victim_primary_job']}",
    axis=1
)
df1['source_data'] = 'df1'
df2['source_data'] = 'df2'
df3['source_data'] = 'df3'
```

```python
base_cols = df1.columns.tolist()
for col in base_cols:
    if col not in df2.columns:
        df2[col] = np.nan
    if col not in df3.columns:
        df3[col] = np.nan
```

```python
extra_cols_2 = [col for col in df2.columns if col not in base_cols]
extra_cols_3 = [col for col in df3.columns if col not in base_cols]

for col in extra_cols_2:
    if col not in df1.columns:
        df1[col] = np.nan
for col in extra_cols_3:
    if col not in df1.columns:
        df1[col] = np.nan
```

```python
merged_df = pd.concat([df1, df2, df3], 
                      axis=0, 
                      ignore_index=True)
columns_to_drop = [
    "event_date", "facility_type", "occupational_category_of_person_affected", "aggressor", "type_of_violence",
    "primary_assault_description", "primary_contributing_factors", "severity_of_assault",
    "emotional_and__or_psychological_impact", "level_of_care_needed", "response_action_taken",
    "mha_facility_type", "wpv_job_role", "wpv_class_of_individual", "wpv_event_type",
    "wpv_type_of_violence_wpv", "wpv_org_contributing_factors", "wpv_psych_impact",
    "injury_assessed_by", "wpv_staff_immediate_action", "event_month", "event_year",
    "victim_primary_job", "days_missed", "source"
]

merged_df = merged_df.drop(columns=columns_to_drop, errors="ignore")
merged_df.head()
```

```python
if 'event_time' in merged_df.columns:
    merged_df['event_time'] = pd.to_datetime(merged_df['event_time'], errors='coerce')

categorical_cols = ['severity', 'violence_type', 'victim_profession', 
                    'department', 'perpetrator_type']
for col in categorical_cols:
    if col in merged_df.columns:
        merged_df[col] = merged_df[col].astype(str).str.strip()

dedup_cols = ['event_time', 'victim_profession', 'department', 'assault_desc']
merged_df.drop_duplicates(subset=dedup_cols, keep='first', inplace=True)
```

```python
merged_df['victim_profession'] = merged_df['victim_profession'].fillna("Unknown")
merged_df['department'] = merged_df['department'].fillna("Unknown")
df = merged_df.copy()
```

```python
def clean_violence_type(val):
    if pd.isna(val):
        return "Unknown"
    val = str(val).lower()
    if "verbal" in val and "physical" in val:
        return "both"
    elif "verbal" in val:
        return "verbal"
    elif "physical" in val:
        return "physical"
    elif "threat" in val:
        return "threat"
    else:
        return "other"

df["violence_type"] = df["violence_type"].apply(clean_violence_type)

def clean_severity(val):
    if pd.isna(val):
        return "Unknown"
    val = val.lower()
    if "mild" in val:
        return "Mild"
    elif "moderate" in val:
        return "Moderate"
    elif "severe" in val:
        return "Severe"
    elif "none" in val:
        return "None"
    else:
        return "Unknown"

df["severity"] = df["severity"].apply(clean_severity)

def clean_perp_type(val):
    if pd.isna(val):
        return "Unknown"
    val = val.lower()
    if "inpatient" in val or "outpatient" in val or "resident" in val or "patient" in val:
        return "patient"
    elif "employee" in val:
        return "employee"
    elif "visitor" in val:
        return "visitor"
    elif "affiliate" in val:
        return "affiliate"
    else:
        return "other"

df["perpetrator_type"] = df["perpetrator_type"].apply(clean_perp_type)

def clean_profession(val):
    if pd.isna(val):
        return "Unknown"
    val = val.lower()
    if "nurse" in val:
        return "nurse"
    elif "tech" in val or "technician" in val:
        return "tech"
    elif "admin" in val or "coordinator" in val or "support" in val:
        return "admin"
    elif "security" in val:
        return "security"
    elif "therapist" in val:
        return "therapist"
    else:
        return "other"

df["victim_profession"] = df["victim_profession"].apply(clean_profession)


df["department"] = df["department"].fillna("unknown").str.strip().str.lower()


categorical_fields = ["violence_type", "severity", "perpetrator_type", "victim_profession", "department"]
df[categorical_fields] = df[categorical_fields].fillna("Unknown")

df["department"] = df["department"].replace(
    to_replace=r"(?i)^ed\b.*",
    value="emergency department",
    regex=True
)

df.to_csv("all_merged_cleaned_standardized.csv", index=False)
```
